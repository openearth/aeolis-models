```python
%pylab inline
import netCDF4
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pickle
from mpl_toolkits.axes_grid1 import make_axes_locatable


root = r'D:\Documents\GIT_checkouts\aeolis-models\sandmotor'

# load colormap
fp = open('cmap.pkl', 'rb')
shore_map = pickle.load(fp)
fp.close()
```

```python
n = 20 # number of time steps

with netCDF4.Dataset(os.path.join(root, 'aeolis_10year_nothing.nc'), 'r') as ds:
    t = ds.variables['time'][:] / 3600.
    x = ds.variables['x'][:,:]
    y = ds.variables['y'][:,:]
    zb = ds.variables['zb'][:]

    nx = x.shape[1]
    ny = y.shape[0]


    #ix = np.isinf(data)
    #data[ix] = np.nan
#    f, axarr = plt.subplots(2, sharex=True)
 #   axarr[0].plot(x, y)
  #  axarr[0].set_title('Sharing X axis')
  #  axarr[1].scatter(x, y)
    #initialize frame
    fig, axs = plt.subplots(2,figsize=(10,9), facecolor='white')
    axs[0].set_aspect('equal')
    axs[0].set_title('Sand Motor Morphology')
    axs[0].set_xlabel('Along-shore [m]')
    axs[0].set_ylabel('Cross-shore [m]')
    p = axs[0].pcolor(y, x, zb[0,...], cmap=shore_map, vmin=-12, vmax=8)
    div = make_axes_locatable(axs[0])
    cax = div.append_axes("right", size="4%", pad=0.15)
    cbar = plt.colorbar(p, cax=cax)#, ticks=np.arange(0,10,.1), format="%.2g")

    axs[1].set_aspect('equal')
    axs[1].set_title('Predicted Morphology Change (after 30 days)')
    axs[1].set_xlabel('Along-shore [m]')
    axs[1].set_ylabel('Cross-shore [m]')
    p = axs[1].pcolor(y, x, zb[1,...]-zb[0,...], cmap='bwr', vmin=-0.5, vmax=.5)
    div = make_axes_locatable(axs[1])
    cax = div.append_axes("right", size="4%", pad=0.15)
    cbar = plt.colorbar(p, cax=cax)#, ticks=np.arange(0,10,.1), format="%.2g")


#    plt.colorbar(p)
#    fig.show()
```
